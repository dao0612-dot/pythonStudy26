```python
#문자열의 길이를 구하는 len() 함수가 있다.
#입력된 주민번호 글자수를 알아내는 코드
#id/pw에 대한 글자수 검증용으로 활용할 수 있다.

inputSsn = "120604-4345678"
type(inputSsn)
```

```python
#주민번호 앞 2자리를 생년으로 추출 year
#주민번호 3~4 자리를 생월로 추출 mounth
#주민번호 5~6 자리를 생일로 추출 day
#주민번호 8번째 1자리를 남여 추출 sex
```

```python
year = inputSsn[0:2]
mounth = inputSsn[2:4]
day = inputSsn[4:6]
sex = "남" if int(inputSsn[7]) % 2 == 1 else "여"
```

```python
print("귀하에 생년은" + year + "년",
"귀하에 생월은" + mounth + "월",
"귀하에 생일은" + day + "일 입니다.", sex)
```

```python
#주민번호를 이용해서 남여 구분을 위한 코드
#문자열로 인식된 숫자를 숫자로 계산시 타입변환 필요
ssnNumber = inputSsn[7]
#print(ssnNumber)
#type(ssnNumber)
ssnNumber = int(ssnNumber) #문자가 숫자로 타입변환진행 int(문자)
type(ssnNumber)
if ssnNumber % 2 == 1 : #숫자로된 성별이 2로 나눈 나머지가 1?
    sex = "남자"
else:
    sex = "여자"
print(sex)
```

```python
#미션 주민번호 8번째 글자가 1,2,5,6이면 1900년 생
#3,4,7,8이면 2000년생으로 만들어보세요!!!

if ssnNumber == 1 :
    yearif = "19" + year #문자인 경우 + 기호는 붙임
elif ssnNumber == 2 :
    yearif = "19" + year
elif ssnNumber == 5 :
    yearif = "19" + year
elif ssnNumber == 6 :
    yearif = "19" + year
elif ssnNumber == 9 :
    print("당신은 외계인입니다. 돌아가세요!!!")
else:
    yearif = "20" + year

print(yearif)
```
