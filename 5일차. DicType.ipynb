```python
#딕셔너리 타입은 key : value로 이루어진다.
#{key1 : value1,key2 : value2 ......}
#대부분 웹 프로그래밍 언어는 json이라는 값으로 처리하는데
#유사한 형태이다.


dic = {"name" : "kkw","age" : 40}
#        k        v     k     v

type(dic)
print(dic["name"])
#           k
```

```python
a = {1 : 'hi'}
print(a[1])
#       k
```

```python
b = {"a" : [1,2,3]}
#     k       v

print(b["a"][1])
```

```python
print(dic)
```

```python
#dic 변수에 email key를 갖는 값을 추가하는 방법
dic["email"] = "mbc@kkw.com"
print(dic)
```

```python
#dic 변수에 age를 삭제해보자.
del dic['age']
print(dic)
```

```python
dic['name'] = '김기원'
print(dic)
```

```python
#key의 리스트를 생성하려면 : a.keys()
dic.keys()
```

```python
#key의 값들을 리스트로 생성하려면
dic.values()
```

```python
#key와 value의 쌍으로 튜플처리하려면
dic.items()
```

```python
#get을 이용해 키를 넣으면 값이 나오도록
dic.get("name")
```

```python
#key에 해당하는 값을 가져올 때 없는 key를 넣으면 에러발생?
dic["age"] #없는 키를 가져오려면 오류발생 keyerror
#이것은 프로그램 실행중 치명적인 오류로 강제 종료 된다.
```
