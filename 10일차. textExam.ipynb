```python
#파일로 더미데이터 및 추가자료를 관리해본다.
#지금까지는 변수 선언할 때 초기값을 리스트에 담아 제공 받았다.
#그러나 프로그램을 종료하면 추가한 값은 사라진다!!!!
#자료를 영구적으로 보관하는 방법은 파일로 저장하던지 데이터베이스에 보관

#서버 : 24시간 365일 동안 멈춤없이 서비스를 제공하는 컴퓨터
#서비스 : 웹서비스, 게임서비스, 쇼핑서비스, 은행뱅킹서비스....

#파일 저장 : 텍스트파일(메모장), csv(엑셀), json(웹api) 등으로 저장이 가능하다.
```

```python
f = open("c:/mbc/새파일.txt","w")
for i in range(1, 11) : #range(1~10까지 값을 가지고 있다)
    data = "%d번째 줄입니다. \n" %i
    #         정수처리      엔터  for i
    f.write(data) #.write 기록!!!
    #파일에 저장
f.close()
```

```python
#파일에 있는 내용을 가져와 보자!!!!
# .readline() 파일에 있는 내용 줄(행 : 가로)단위로 가져오기
f = open("c:/mbc/새파일.txt",'r') #r은 읽기 전용 모드
line = f.readline() #기본값은 첫줄 가져오기
print(line)
f.close()
```

```python
#모든 줄을 읽어 화면에 출력하고 싶으면
f = open("c:/mbc/새파일.txt",'r')
while True: # False 처리될때까지 반복분
    line = f.readline() #1줄씩 읽는다.
    if not line : break #line이 없으면 break로 중지
    print(line)
f.close()
#메모장 내용에 엔터가 적용이 되어있는 상태임 \n \r 라인피드 캐리지리턴
```

```python
# .readlines() 메모장 파일을 리스트로 한번에 불러온다.
# 파일의 모든 줄을 읽어 리스트 형태로 반환
# 반환값은 문자열 리스트 ['첫번째줄\n','두번째줄\n',,,,,,,,]
# 파일 내용을 모두 메모리에 로드하기 때문에 파일크기가 클경우 메모리 부족!
# 장점 : 파일의 모든 줄을 한번에 가져와 처리해야 할 때 편리

f = open("c:/mbc/새파일.txt",'r')
lines = f.readlines() # 파일에 있는 모든 내용을 리스트 처리
#print(lines)
for line in lines:
    print(line)
f.close()
```

```python
#메모장에 엔터효과인 \n 을 지우고 싶으면 strip
#strip은 문자열내에서 원하는 문자열 또는 공백을 제거함
# lstrip() 원하는 문자열 또는 공백(엔터)을 왼쪽에서 제거
# rstrip() 원하는 문자열 또는 공백(엔터)을 오른쪽에서 제거
f = open("c:/mbc/새파일.txt",'r')
lines = f.readlines()
for line in lines:
    line = line.strip() #줄끝에 줄 바꿈 문자를 제거
    print(line)
f.close()
```

```python
#read()를 이용하면 파일 문자열 읽기
#파일의 내용 전체를 문자열로 반환
f = open("c:/mbc/새파일.txt",'r')
data = f.read()
print(data) #for문을 사용하지 않고 전체를 읽어 출력!!!
f.close()
```

```python
#for를 이용해 줄단위로 가져오기
f = open("c:/mbc/새파일.txt",'r')
#파일 객체는 기본적으로 for문과 같이 함께 사용하여 줄단위로 읽기도 가능
for line in f :
    print(line)
f.close()
```

```python
#파일에 새로운 내용을 추가 append -> a
#w로 파일을 열면 이미 존재하는 파일이면 그 내용이 사라지고 덮어 쓰기
#a 모드는 원래 있던 값을 유지하면서 새로운 값을 맨 아래에 추가!!!

f = open("c:/mbc/새파일.txt",'a') #append (추가)
f.write("-------------------------")
for i in range(11,20):
    data = "%d번째 줄입니다. \n" %i
    f.write(data)
f.close()
```
