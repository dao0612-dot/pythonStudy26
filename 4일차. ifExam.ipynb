```python
#if문은 조건문으로 판단하여 참(True) 거짓(False)로 판단하는 코드
#if 조건 :
#   조건이 참일 때 실행하는 문
#elif 조건 :
#   위 조건이 False일때 또 비교하는 조건문
#else:
#   위 조건이 모두 해당하지 않을 때 실행하는 조건문

#학점 계산용 코드
avg = int(input("평균을 입력하세요."))
if avg >= 90 :
    grade = "A" #90점 이상이면 A

elif avg >= 80 :
    grade = "B" #80점 이상이면 B

elif avg >= 70 :
    grade = "C" #70점 이상이면 C

else:
    grade = "F" #그 외엔 전부 F

print(f"귀하의 등급은 {grade}등급입니다.") #.format 도 사용할 줄 알아야한다. f포맷팅은 3.6부터 생긴 기능, 3.2를 사용하는 회사에선 불가능
```

```python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit + 1
    print(f"나무를 {treeHit}번 찍었습니다.")
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

```python
coffee = 3
while True:
    money = int(input("돈을 넣어 주세요 :"))
    if money == 300 :
        coffee = coffee - 1
        print("커피를 줍니다.")

    elif money > 300 :
        coffee = coffee - 1
        print("거스름돈을 %s만큼 줍니다." % (money - 300))

    else:
        print("돈을 돌려주고 커피를 주지 않습니다.")

    if coffee == 0 :
        print("커피가 전부 떨어졌습니다.")
        break
```

```python
marks = [90,25,67,45,80]

number = 0
for mark in marks:
    number = number + 1
    if mark >= 60:
        print(f"{number}번 학생은 합격입니다.")

    else:
        print(f"{number}번 학생은 불합격입니다.")
```

```python
marks = [90,25,67,45,80]
number = 0
for mark in marks:
    number = number + 1
    if mark < 60:
        continue

    else:
        print(f"축하합니다. {number}번 학생은 합격입니다.")
```

```python
marks = [90,25,67,45,80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f"축하합니다. {number + 1}번 학생은 합격입니다.")
```

```python
for i in range(2,10):
    for j in range(1,10):
        print(i*j, end=" ")
    print('') #print()함수는 기본적으로 \n를 내장하고 있기 때문에, 2번째 for문이 끝나서 아래의 print('')가 실행되서 엔터가 쳐진다. 즉 3의 구구단은 다음열에서 쭉 실행된다.
```

```python
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
print(result)
```

```python
a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```
