```python
#함수는 반복적인 코드를 위쪽에 선언하여 호출함으로써
#프로그램을 가독성 좋고 간편하게 구현할 수 있다.

#클래스가 없는 기능을 함수라고 부르고
#클래스가 있는 기능을 메서드라고 한다.

#def 는 함수는 메서드를 만드는 예약어로
#def 함수명(매개변수 == 인수 == 파라미터) :
#                                들여쓰기 필수
#   함수 내부의 실행문들......
#   return값 = 함수 종료시 호출한 곳으로 넘겨주는 값
#들여쓰기를 끝내면 함수 종료

#변수 = 함수명(매개변수 == 인수 == 파라미터)
#위에서 만든 함수를 호출하여 처리 후 리턴을 제공 받는다.
```

```python
from textwrap import dedent


#함수를 만든다.(함수를 정의한다.)
def add(a,b): #add() 함수를 만들고 인수는 a와 b를 만든다.
    return a+b    #받은 a,b를 더하고 돌려준다.
```

```python
#위에서 만든 함수를 활용해보자
c = add(5,10) #위에서 만든 add 함수에 5와 10을 넣고 5 + 10의 리턴(결과)를 c에 넣는다.
print(c)
```

```python
def subMenu(select):
    if select == "1":
        print(dedent(f"""
        ====회원관리====
        1. 회원가입
        2. 로그인
        3. 회원정보 수정
        """))

    elif select == "2":
        print("====성적관리====")
        print("1. 성적등록")
        print("2. 성적보기")
        print("3. 성적수정")
        print("4. 성적백업")
        print("5. 상위메뉴로")
```

```python
subMenu(input("1~5사이값 입력"))
```

```python
#만약 입력값이 몇개가 될지 모를 때는??
def add_many(*args): # *을 붙이면 몇개의 값이 들어오든!!!
# *의 뜻은 와일드 문자로 김씨로 시작한다.~ "김*"
# "*원" 이면 원으로 끝나는 문자
# "*딸기*" 딸기라는 글자가 포함된!
#args는 인수라는 뜻으로 다른 글자로 사용해도 됨!
#그러나 프로그래머들의 관례로 대부분 args로 사용함
    result = 0 #결과를 보관하는 변수
    for i in args: #args는 튜플로 들어옴
        result = result + i #튜플로 되어 있는 값을 번갈아 더함
    return result
```

```python
#add_many 테스트
print(add_many(1,2,3,))
```

```python
resultText = add_many("안녕하세요","반갑습니다.","파이썬 지옥입니다.")
print(resultText)
```
