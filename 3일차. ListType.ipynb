```python
#리스트에 C R U D 기법을 적용해보자.
#CREATE -> 생성, 추가, 삽입 .append()
#READ   -> 전체리스트보기, 1개 자료보기 .index()
#UPDATE -> 자료수정  passwords[3] = "4486"
#DELDTE -> 자료삭제  del passwords[3] .remove(3)  .pop(3)
```

```python
#리스트에 자료를 추가해보자!!!
ids = ["kkw","lhf","ksb"] #회원 id
ids.append("MBC")
print(ids)
```

```python
passwords = ["1234","4321","4567"]
passwords.append("8888")
print(passwords)
```

```python
#리스트에 있는 값 가져오기
ids.index("MBC")
```

```python
#MBC ID에 인덱스를 알아봐 패스워드의 값을 찾아온다.
passwords[3]
```

```python
#알아온 password 값을 4486으로 수정해본다.
passwords[3] = "4486"
print(passwords)
```

```python
#3번째 인덱스에 암호를 삭제해본다.
passwords.remove("4486")
print(passwords)
```

```python
#pop은 삭제되는 내용을 출력하고 삭제한다.
#passwords.pop() -> 맨 마지막 자료를 삭제
#passwords.pop(2) -> 2 인덱스를 삭제한다.
passwords.pop(2)
```
